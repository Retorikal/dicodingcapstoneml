```python
df_user = pd.read_csv('/home/moseshubert/Documents/bangkit_project/dataset/course_recommender_dataset/user_activities.csv')
df_user.head()
```

```python
df_user.tail()
```

```python
df_user.info()
```

```python
df_user_stack = df_user.assign(on_progress_courses=df_user['on_progress_courses'].str.split(",")).explode("on_progress_courses")
df_user_stack.head()
```

```python
df_user_stack.info()
```

```python
df_user_stack['graduated_courses'].unique()
```

```python
df_user_full_stack = df_user_stack.assign(graduated_courses=df_user_stack['graduated_courses'].str.split(",")).explode("graduated_courses")
df_user_full_stack
```

```python
df_user_full_stack.describe(include='object')
```

```python
df_user_full_stack['on_progress_courses'].unique()
```

```python
df_user_full_stack['graduated_courses'].unique()
```

```python
df_gc = df_user_full_stack[['id','graduated_courses']]
df_gc.head()
```

```python
df_opc = df_user_full_stack[['id','on_progress_courses']]
df_opc['on_progress_courses'].unique()
```

```python
# df.rename(columns = {'old_col1':'new_col1', 'old_col2':'new_col2'}, inplace = True)
df_gc.rename(columns={'graduated_courses':'course_name'}, inplace=True)
df_gc.head()
```

```python
# df.rename(columns = {'old_col1':'new_col1', 'old_col2':'new_col2'}, inplace = True)
df_opc.rename(columns= {'on_progress_courses': 'course_name'}, inplace=True)
df_opc.head()
```

```python
# df = pd.concat([df, df2], ignore_index = True, axis = 0)
df_concat = pd.concat([df_gc, df_opc], ignore_index=True, axis=0)
df_concat.info()
```

```python
df_concat.nunique()
```

```python
df = df_concat.dropna()
df.nunique()
```

```python
df.info()
```

```python
df.describe(include='object')
```

```python
df['course_name'].unique()
```

```python
df.to_csv('/home/moseshubert/Documents/bangkit_project/dataset/course_recommender_dataset/users.csv', index=False)
```
